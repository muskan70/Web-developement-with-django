## Beautiful Soup Code

```python
import bs4
import requests
url="https://www.passiton.com/inspirational-quotes"
```

```python
response=requests.get(url)
response.content
```

```python
#HTML parser
soup=bs4.BeautifulSoup(response.content)
```

```python
article_element=soup.findAll('div',**{'class':"col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top"})
print(type(article_element))
```

```python
print(article_element[0])
print(len(article_element))
```

```python
article=article_element[0]
article.img.attrs['src']
```

```python
with open('img/inspiration.jpg','wb') as file:
    img_url=article.img.attrs['src']
    response=requests.get(img_url)
    file.write(response.content)
```

```python
for i,article in enumerate(article_element):
    with open('img/inspiration{}.jpg'.format(i),'wb') as file:
        img_url=article.img.attrs['src']
        response=requests.get(img_url)
        file.write(response.content)
```
